## Importar y generar reporte


```python
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
# Cargar dataset
df = pd.read_csv('Evaluaciones_Agropecuarias_Municipales_EVA_20251102.csv')
```

```python
# Generar reporte
profile = ProfileReport(df, title="Perfil del Dataset: Rendimiento de Cultivos en Colombia", explorative=True)
```

```python
# Guardar reporte HTML
profile.to_file("perfil_rendimiento_cultivos.html")
```
